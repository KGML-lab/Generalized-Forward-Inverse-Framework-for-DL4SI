```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from plot_settings import global_setting
from matplotlib.lines import Line2D
global_setting()
```

```python
model_names = ['FlatVel-A', 'FlatVel-B',
              'CurveVel-A', 'CurveVel-B',
              'FlatFault-A', 'FlatFault-B',
              'CurveFault-A', 'CurveFault-B',
              'Style-A', 'Style-B']

metrics = ["MSE", "MAE", "SSIM"]
```

```python
results_dict = {}
for metric in metrics:
    results_dict[metric] = pd.DataFrame(columns=["Dataset", "Metric"])
    results_dict[metric]["Dataset"] = model_names
    results_dict[metric]["Metric"] = metric
```

```python
results_dict["MSE"]
```

```python
model_dict = {"InversionNet (L1+L2+VGG)": "InversionNet",
              "InversionNet": "InversionNet_ckpt", #InversionNet (L1)
              "VelocityGAN": "Velocity_GAN", # VelocityGAN (L1)
              "Invertible XNet (Adam)" : "Invertible_XNet_Adam",
              "Invertible X-Net" : "Invertible_XNet", # "Invertible XNet (Adamax)"
              "Invertible X-Net (Cycle)" : "Invertible_XNet_cycle_warmup", # "Invertible XNet (Adamax Cycle)"
              "IUnetInverseModel" : "IUnetInverseModel",
              "Latent U-Net (Small)" : "UNetInverseModel_17M",
              "Latent U-Net (Large)" : "UNetInverseModel_33M",  
              "Auto-Linear": "AutoLinear_Inversion_ckpt"
             }

color_dict = {
            "InversionNet (L1+L2+VGG)": 'lightcoral',           # Blue
            "InversionNet": 'orange',                 # Green
            "VelocityGAN": 'red',                    # Red
            "XNet (Adam)": 'blue',                        # Cyan
            "Invertible X-Net": 'lightskyblue',                   # Magenta
            "Invertible X-Net (Cycle)": 'steelblue',              # Yellow
            "IUnetInverseModel": 'black',                 # Black
            "Latent U-Net (Small)": 'forestgreen',              # Purple
            "Latent U-Net (Large)": 'limegreen' ,              # Orange
            "Auto-Linear": 'chocolate'
        }

mapping = {
    'FlatVel-A': 'FVA',
    'FlatVel-B': 'FVB',
    'CurveVel-A': 'CVA',
    'CurveVel-B': 'CVB',
    'FlatFault-A': 'FFA',
    'FlatFault-B': 'FFB',
    'CurveFault-A': 'CFA',
    'CurveFault-B': 'CFB',
    'Style-A': 'STA',
    'Style-B': 'STB'
}

basepath = "../Metrics_final/"
prefix = "eval_metric"
mode = "Velocity"

sheets = ["Velocity_unnorm_MAE", "Velocity_unnorm_MSE", "Velocity_SSIM"]
```

```python
for key, value in model_dict.items():
    grid_file = os.path.join(basepath, prefix+value+".xlsx")
    model_name = key
    for metric in sheets:
        metric_type = metric.split("_")[-1]
        try:
        #read the particular metric sheet in the result grid file (dataset vs dataset)
            df = pd.read_excel(grid_file, engine='openpyxl', header=0, index_col=0, sheet_name=metric)
            metric_diag = [df[dataset.lower()][dataset] for dataset in model_names]
            results_dict[metric_type][model_name] = metric_diag
        except:
            results_dict[metric_type][model_name] = [None]*len(results_dict["MSE"])
```

```python
results_dict["MSE"]
```

```python
for metrics in results_dict.keys():
    results_df = results_dict[metrics]
    results_df["Dataset"] = [mapping[dataset] for dataset in results_df["Dataset"]]
```

# Inverse Plot

```python
results_mse = results_dict["MSE"]

arch_names_xnets = ["XNet (Adam)","XNet (Adamax)","XNet (Adamax Cycle)"]
color_list = [color_dict[arch_name] for arch_name in arch_names_xnets]

plt.figure(figsize=(12, 2))
ax = results_mse.plot(x="Dataset",y=arch_names_xnets, kind="bar",figsize=(20, 3), rot=0, fontsize=14, color=color_list)
plt.yscale("log")
ax.set_ylabel("MSE", fontsize=14)
ax.set_xlabel("Dataset", fontsize=14)


ax.set_ylim([100, 1000000])
plt.legend(loc="upper center", ncol=4, fontsize=12)
plt.show()
```
